```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Lists of tickers
tickers_水産農林業 = ["1301.T", "1332.T", "1333.T", "1375.T", "1376.T", "1377.T", "1379.T", "1380.T", "1381.T", "1382.T", "1383.T", "1384.T"]
tickers_鉱業 = ["1514.T", "1515.T", "1518.T", "1605.T", "1662.T", "1663.T"]

all_tickers = tickers_水産農林業 + tickers_鉱業

# Fetch YTD data and market capitalizations for all tickers
data = yf.download(all_tickers, start="2023-01-01", end="2023-12-31", group_by='ticker')
info = {ticker: yf.Ticker(ticker).info for ticker in all_tickers}

# Extract market capitalizations
market_caps = {ticker: info[ticker]['marketCap'] for ticker in all_tickers if 'marketCap' in info[ticker]}

def compute_index(tickers_list):
    index_data = pd.DataFrame()
    total_market_cap = sum(market_caps[ticker] for ticker in tickers_list if ticker in market_caps)
    for ticker in tickers_list:
        if ticker in data and ticker in market_caps:
            index_data[ticker] = data[ticker]['Adj Close'] * (market_caps[ticker] / total_market_cap)
    index_data['Index'] = index_data.sum(axis=1)
    return index_data['Index']

# Compute indices
index_水産農林業 = compute_index(tickers_水産農林業)
index_鉱業 = compute_index(tickers_鉱業)

# Plot the indices
plt.figure(figsize=(14, 7))
index_水産農林業.plot(label="水産農林業 Index")
index_鉱業.plot(label="鉱業 Index")
plt.title("Indices YTD Performance")
plt.xlabel("Date")
plt.ylabel("Index Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the list of tickers for the "石油石炭製品" industry
ticker_石油石炭製品 = ['3315.T', '5010.T', '5011.T', '5013.T', '5015.T', '5017.T', '5018.T', '5019.T', '5020.T', '5021.T']

# Create an empty DataFrame to store market cap data
market_cap_data = pd.DataFrame()

# Define the YTD date range
start_date = '2023-01-01'
end_date = '2023-12-31'

# Loop through each date and calculate the total market cap for each stock
for date in pd.date_range(start=start_date, end=end_date):
    date_str = date.strftime('%Y-%m-%d')
    market_caps = []

    for ticker in ticker_石油石炭製品:
        # Download stock info for the ticker
        info = yf.Ticker(ticker)
        hist_data = info.history(period="1d", interval="1d", start=date_str, end=date_str)

        if not hist_data.empty:
            market_cap = hist_data['MarketCap'].values[0]
            market_caps.append(market_cap)

    # Create a dictionary with date and market cap values for each stock
    market_cap_dict = {'Date': date, **{ticker: market_cap for ticker, market_cap in zip(ticker_石油石炭製品, market_caps)}}

    # Append the data for the current date to the market_cap_data DataFrame
    market_cap_data = market_cap_data.append(market_cap_dict, ignore_index=True)

# Set the 'Date' column as the index
market_cap_data.set_index('Date', inplace=True)

# Calculate the total market cap for each day by summing the market caps of all stocks
market_cap_data['TotalMarketCap'] = market_cap_data.sum(axis=1)

# Plot the line chart of the TotalMarketCap historical data
plt.figure(figsize=(12, 6))
plt.plot(market_cap_data.index, market_cap_data['TotalMarketCap'], label='Total Market Cap')
plt.title('Total Market Cap Historical Data')
plt.xlabel('Date')
plt.ylabel('Total Market Cap')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
